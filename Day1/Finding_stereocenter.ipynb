```python
from IPython.core.display import Image
from rdkit.Chem import rdCIPLabeler


#@title Load molecule { run: "auto" }
#@markdown Here we will load the molecule using SMILES representation. The SMILES representation can be generated at https://pubchem.ncbi.nlm.nih.gov//edit3/index.html. Copy the smiles text into this cell. This is simple coding and does not use any advanced computational chemistry techniques like DFT.

smiles = "" #@param {type:"string"}


# Establish RDKit mol object
rd_mol = Chem.MolFromSmiles(smiles)
rdmol_hs = Chem.rdmolops.AddHs(rd_mol)
rdCIPLabeler.AssignCIPLabels(rdmol_hs)

MolsToGridImage([rdmol_hs],subImgSize=(500,500))
```
